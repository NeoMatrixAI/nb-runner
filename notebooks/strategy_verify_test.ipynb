# Get Data

```python
####  Don't touch this shell ####

import requests
import pandas as pd

def fetch_candlestick_data_coinapi(data_apikey, symbol, timeframe='1min', startDate=None, endDate=None, tradeType='future'):

    # future URL
    if tradeType == 'future':
      url = f'https://crypto.fin.cloud.ainode.ai/{data_apikey}/get/candlestick/future/{timeframe}?exchange_nm=bitget&symbol={symbol}&startDate={startDate}'

    # Add to URL if endDate is not None
    if endDate is not None:
        url += f'&endDate={endDate}'
    df = pd.DataFrame(requests.get(url).json()['data'])
    df.drop(['exchange_nm','symbol'], axis = 1, inplace = True)
    df['date'] = pd.to_datetime(df['date']).dt.strftime('%Y-%m-%d %H:%M:%S')
    df = df.set_index('date')
    return df
```

```python
from datetime import datetime, timedelta

#### Enter the data API key provided
data_apikey = "Input your Data API Key"
#####

#### Don't touch
symbols = ['BTCUSDT', 'ETHUSDT', 'XRPUSDT', 'EOSUSDT', 'BCHUSDT', 'LTCUSDT', 'ADAUSDT', 'ETCUSDT', 'TRXUSDT', 'DOTUSDT'] # Fixed variable for verify test
timeframe = '1min' # Current fixed variable
startDate = (datetime.now().date() - timedelta(days=2)).strftime('%Y-%m-%d') # Fixed variable for verify test
endDate = (datetime.now().date() - timedelta(days=1)).strftime('%Y-%m-%d') # Fixed variable for verify test
tradeType = 'future' # Current fixed variable

df = pd.DataFrame()
for symbol in symbols:
    new = fetch_candlestick_data_coinapi(data_apikey, symbol, timeframe, startDate, endDate, tradeType)
    df = pd.concat([df,new.close], axis = 1)

df.columns = symbols
df
```
